```python
#imprtant libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import KFold
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
import os
```

```python
#directories
orginal_dir='/home/usama/Desktop/obd2_project'
data_dir='/home/usama/Desktop/obd2_project/OBD II Dataset'
```

```python
#import csv directory folder
dataset=os.path.join(data_dir,"Modified_OBDII.csv")
```

```python
#read csv file
dataframe=pd.read_csv(dataset)
```

```python
#set the values
data_set=dataframe.values
```

```python
#apply get dummies to convert string data into binary form
dataset=pd.get_dummies(dataframe,prefix_sep='_',drop_first=False)
dataset.head()
```

```python
k=5
k_means=KMeans(n_clusters=k,init='k-means++',n_init=12)
k_means.fit(dataset)
```

```python
label=k_means.labels_
```

```python
dataset['Outcomes']=label
```

```python
dataset.to_csv(r'/home/usama/Desktop/obd2_project/OBD II Dataset/complete_labeled_OBDII.csv',index=False)
```
